```python
import sys
sys.path.append("../densities")
from Gauss import Gauss
from inverse_transform_sampling import it_sampling
import numpy as np
import matplotlib.pyplot as plt
```

### Compute target density

```python
g = Gauss([0], [1])
x_vec  = np.linspace(-3., 3., 1000)
points = g.evaluate(x_vec)
```

```python
fig, ax = plt.subplots(1,1, figsize=(5,3))
ax.plot(x_vec, points)
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$p_X(x)$')
```

### Do inverse transform sampling and compare histogram

```python
samples = it_sampling(g, num_samples=10000, x_intervals=1000)
```

```python
fig, ax = plt.subplots(1,1, figsize=(5,3))
ax.hist(samples, bins='auto')
ax.set_xlabel(r'$x$')
ax.set_ylabel('histogram count')
```
